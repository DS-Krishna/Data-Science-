Top 20 Days Price of the ACLBSL

```python
import pandas as pd
import numpy as np
df = pd.read_csv("/content/ACLBSL_subset.csv")
top20_df = df.tail(20).reset_index()
top20_df
```

Coverting the Columns Volume to the list

```python
top20_volume = top20_df['volume'].tolist()
top20_volume
```

Displaying the Min, Max, and the Mean of the Volume

```python
maximum_volume = top20_df['volume'].max()
minimum_volume = top20_df['volume'].min()
mean_volume = sum(top20_volume)/len(top20_volume)
print(f"Maximum Volume: {maximum_volume}")
print(f"Minimum Volume: {minimum_volume}")
print(f"Mean Volume: {mean_volume}")
```

Calculating the Range of the Volume

```python
range = maximum_volume - minimum_volume
print(range)
```

Variance, Standard Deviation Calculation

```python
sum = 0
for i in top20_volume:
  sum += (i - mean_volume)**2
variance = sum / (len(top20_volume)-1)
print(f"Variance: {variance}")
standard_deviation = variance ** 0.5
print(f"Standard Deviation: {standard_deviation}")
```

Z score Value Calculation ::

```python
z_score = (top20_volume[-1] - mean_volume)/standard_deviation
print(z_score)
```
